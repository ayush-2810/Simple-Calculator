```python
from tkinter import *   #Importing tkinter library
expression=''   #Declaring global variable

def entry(x):   #Function to update the entries
    global expression
    expression = expression + str(x)   #Concanating inputs for showing on entry bar
    calculations.set(expression)   #Update calculations
    
def equal():   #Function for showing the output
    try:
        global expression
        answer= str(eval(expression))   #Evaluating answer
        calculations.set(answer)   
        expression=answer
        
    except:
        ans='Error'   
        calculations.set(ans)

def clear():   #Function for clearing the screen
    global expression
    expression = ''
    calculations.set(expression)
    
def delt():   #Delete function 
    global expression
    expression=expression[:-1]   #Removes last character
    calculations.set(expression)

root = Tk()


root.title('Simple Calculator')   #Title of window
root.geometry('240x200')   #Size of window
root.config(bg='yellow')   #Background colour of window

calculations = StringVar()
entrybox = Entry(root,textvariable=calculations,bg='light grey',width=30)   #Entry bar
entrybox.pack(ipady=8)


ff = Frame(root)   #First frame
clear = Button(ff,text='C',height=1,width=11,command=clear,bg='light grey')   #Clear button
clear.grid(row=0,column=0,sticky=W,padx=3,pady=3)
delete = Button(ff,text='Del',height=1,width=4,command=delt,bg='light grey')  #Delete button
delete.grid(row=0,column=1,sticky=W,padx=3,pady=3)
divide = Button(ff,text='/',height=1,width=4,command=lambda: entry('/'),bg='light grey')
divide.grid(row=0,column=2,sticky=W,padx=3,pady=3)
ff.pack()

ft = Frame(root)   #Top frame 

b7 = Button(ft,text='7',height=1,width=4,command=lambda: entry(7),bg='light grey')   #Button for integer 7
b7.grid(row=1,column=0,sticky=W,padx=3,pady=3)
b8 = Button(ft,text='8',height=1,width=4,command=lambda: entry(8),bg='light grey')
b8.grid(row=1,column=1,sticky=W,padx=3,pady=3)
b9 = Button(ft,text='9',height=1,width=4,command=lambda: entry(9),bg='light grey')
b9.grid(row=1,column=2,sticky=W,padx=3,pady=3)
multiply = Button(ft,text='x',height=1,width=5,command=lambda: entry('*'),bg='light grey')
multiply.grid(row=1,column=3,sticky=W,padx=3,pady=3)
ft.pack()

fm = Frame(root)   #Middle frame

b4 = Button(fm,text='4',height=1,width=4,command=lambda: entry(4),bg='light grey')
b4.grid(row=2,column=0,sticky=W,padx=3,pady=3)
b5 = Button(fm,text='5',height=1,width=4,command=lambda: entry(5),bg='light grey')
b5.grid(row=2,column=1,sticky=W,padx=3,pady=3)
b6 = Button(fm,text='6',height=1,width=4,command=lambda: entry(6),bg='light grey')
b6.grid(row=2,column=2,sticky=W,padx=3,pady=3)
subtract = Button(fm,text='-',height=1,width=5,command=lambda: entry('-'),bg='light grey')
subtract.grid(row=2,column=3,sticky=W,padx=3,pady=3)
fm.pack()

fb = Frame(root)   #Bottom frame

b1 = Button(fb,text='1',height=1,width=4,command=lambda: entry(1),bg='light grey')
b1.grid(row=3,column=0,sticky=W,padx=3,pady=3)
b2 = Button(fb,text='2',height=1,width=4,command=lambda: entry(2),bg='light grey')
b2.grid(row=3,column=1,sticky=W,padx=3,pady=3)
b3 = Button(fb,text='3',height=1,width=4,command=lambda: entry(3),bg='light grey')
b3.grid(row=3,column=2,sticky=W,padx=3,pady=3)
addition = Button(fb,text='+',height=1,width=5,command=lambda: entry('+'),bg='light grey')
addition.grid(row=3,column=3,sticky=W,padx=3,pady=3)
fb.pack()

fl = Frame(root)   #Last frame
b = Button(fl,text='.',height=1,width=4,command=lambda: entry('.'),bg='light grey')
b.grid(row=4,column=1,sticky=W,padx=3,pady=3)
b0 = Button(fl,text='0',height=1,width=4,command=lambda: entry(0),bg='light grey')
b0.grid(row=4,column=2,sticky=W,padx=3,pady=3)
equal = Button(fl,text='=',height=1,width=11,command=equal,bg='light grey')
equal.grid(row=4,column=3,sticky=W,padx=3,pady=3)
fl.pack()



root.mainloop()
```
